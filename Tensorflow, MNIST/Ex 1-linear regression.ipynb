### Understanding linear regression model 

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
```

### Defining input vairables

```python
obs=1000
xz=np.random.uniform(low=-10, high=10, size=(obs,1))
yz=np.random.uniform(-10,10,(obs,1))
input=np.column_stack((xz,yz))
print(input.shape)
```

### Defining noise, target, bias, weights and learning rate

```python
noise=np.random.uniform(-1,1,(obs,1))
```

```python
targets=2*xz+3*yz+5+noise
```

```python
print(targets.shape)
```

```python
init_range=0.1
```

```python
weights=np.random.uniform(-init_range,init_range,(2,1))
```

```python
print(weights.shape)
```

```python
bias=np.random.uniform(-init_range,init_range,(1))
```

```python
bias
```

```python
lr=0.01
```

### Calculating output and loss function by iterations 

```python
for i in range(1000):
    output=np.dot(input,weights) + bias
    delta=output-targets
    loss= np.sum(delta*2)/2/obs
    print(loss)
    delta_scaled=delta/obs
    weights=weights-lr*np.dot(input.T,delta_scaled)
    bias=bias-lr*np.sum(delta_scaled)
```

```python
print(weights,bias)
```

### Plotting the output and target 

```python
plt.plot(output,targets)
plt.xlabel('output')
plt.ylabel('targets')
plt.show()
```
